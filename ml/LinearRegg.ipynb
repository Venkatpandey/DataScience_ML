```python
from sklearn import linear_model
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv')
data.head()
```

```python
plt.scatter(data.year, data['per capita income (US$)'])
```

```python
lg = linear_model.LinearRegression()
year= pd.DataFrame(data.year)
income=pd.DataFrame(data['per capita income (US$)'])
lg.fit(year, income)
```

```python
lg.predict([[2020]])
```
