```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/7_logistic_reg/insurance_data.csv")
data.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(data['age'], data['bought_insurance'], marker='+', color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr, x_t, y_tr, y_t = train_test_split(data[['age']], data.bought_insurance, train_size=0.9)
```

```python
print(x_t)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_tr, y_tr)
```

```python
lr.predict(x_t)
```

```python
lr.score(x_t,y_t)
```

```python
lr.predict_proba(x_t)
```
