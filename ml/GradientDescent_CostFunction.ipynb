```python
import numpy as np
import pandas as pd
import math as ma
from sklearn import linear_model
```

```python
def gradient_descent(x,y):
  m_curr = b_curr = 0
  it = 1000
  n = len(x)
  step = 0.0002
  cost_previous = 0
  for i in range(it):
    y_predict = m_curr * x + b_curr
    cost = (1/n) * sum([val**2 for val in (y-y_predict)])
    if ma.isclose(cost, cost_previous, rel_tol=1e-20):
      break
    cost_previous = cost
    md = -(2/n) * sum(x*(y-y_predict))
    bd = -(2/n) * sum(y-y_predict)
    m_curr = m_curr - step * md
    b_curr = b_curr - step * bd
    print("m {}, b {}, cost {}, it {}".format(m_curr, b_curr, cost, i))
```

```python
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
gradient_descent(x,y)
```
