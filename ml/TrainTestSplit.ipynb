```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/6_train_test_split/carprices.csv")
data.head()
```

```python
x = data[['Mileage', 'Age(yrs)']]
x.head()
```

```python
y = data[['Sell Price($)']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr, x_t, y_tr, y_t = train_test_split(x,y,test_size=0.2)
```

```python
len(x_tr)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_tr, y_tr)
```

```python
lr.predict(x_t)
```

```python
lr.score(x_t,y_t)
```
