```python
from sklearn import linear_model
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/2_linear_reg_multivariate/homeprices.csv")
df.head()
```

```python
median = df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(median)
df.head()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
reg.coef_
reg.intercept_
```

```python
reg.predict([[3000,3,15]])
```

```python
reg.predict([[2500,4,5]])
```

```python
medianT = data['test_score(out of 10)'].mean()
print(medianT)
```

```python
data['test_score(out of 10)'] = data['test_score(out of 10)'].fillna(medianT)
```

```python
linReg = linear_model.LinearRegression()
linReg.fit(data[['experience','test_score(out of 10)','interview_score(out of 10)']], data['salary($)'])
```
